```python
#Cambio la struttura dati per agevolarmi in alcune operazioni

def list_to_dict(lista):
    dizionario = {}
    for row in lista: # Inizializzo il dizionario
        for attr in row.keys():
            dizionario[attr] = []
    for row in lista:
        for attr in row:
            dizionario[attr].append(row[attr])
    return dizionario

def dict_to_list(dizionario):
    lista = []
    first_attribute = list(dizionario.keys())[0]
    n_rows = len(dizionario[first_attribute])
    for i in range(n_rows):
        nuova_riga = {}
        for k in dizionario.keys():
            nuova_riga[k] = dizionario[k][i]
        lista.append(nuova_riga)
    return lista
```

```python
from datetime import datetime #Utilizzo la libreria perche' altrimenti dovrei creare un dizionario che mappa
                              #ogni data ad un giorno della settimana. Ma avrei problemi con gli update nel 
                              #caso in cui il datawarehouse si ingrandisse

Date = []

date_id = 0
for row in dates:
    day = row['date'].split()[0]
    day_of_the_week = datetime.strptime(day, '%Y-%m-%d').strftime('%A')
    month = day[:len(day)-3]
    month_of_the_year = datetime.strptime(day, '%Y-%m-%d').strftime('%B')
    year = month[:len(month)-3]
    quarter_of_the_year = 'Q' + str((int(month.split('-')[1]) - 1) // 3 + 1)
    quarter = year + '-Q' + quarter_of_the_year
    d = {'date_id' : date_id, 
         'day' : day, 
         'day_of_the_week' : day_of_the_week,
         'month' : month, 
         'month_of_the_year' : month_of_the_year,
         'year' : year, 
         'quarter_of_the_year' : quarter_of_the_year,
         'quarter' : quarter}
    Date.append(d)
    date_id += 1
```

### Mappatura tramite la libreria geopy

```python
# Questa cella ci mette 13/14 ore. Io l'ho lanciata in precedenza su un altro notebook creando il file coordinates.txt
# per salvare il valori geografici mappati da geopy. Per questo motivo questa cella non ha output.

from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderUnavailable
from IPython.display import clear_output
import time

coordinate = []
for row in Police:
    t = (row['latitude'], row['longitude'])
    coordinate.append(t)

s = set(coordinate)

geolocator = Nominatim(user_agent="geo_app")
g_id = 0
with open("..\data\coordinates.txt", "w", encoding="utf-8") as file:
    file.write("geo_id,latitude,longitude,state,county,city\n")
    for coordinate in s:
        latitude = coordinate[0]
        longitude = coordinate[1]
        while True:
            try:
                location = geolocator.reverse((latitude, longitude), exactly_one=True)
                indirizzo_parti = location.raw.get("address", {})
                if 'city' in indirizzo_parti:
                    city = indirizzo_parti.get("city", "N/D")
                elif 'town' in indirizzo_parti:
                    city = indirizzo_parti.get("town", "N/D")
                elif 'village' in indirizzo_parti:
                    city = indirizzo_parti.get("village", "N/D")
                elif 'borough' in indirizzo_parti:
                    city = indirizzo_parti.get("borough", "N/D")
                state = indirizzo_parti.get("state", "N/D")
                county = indirizzo_parti.get("county", "N/D")
                result = f"{g_id},{latitude},{longitude},{state},{county},{city}\n"
                file.write(result)
                g_id += 1
                clear_output(wait=True)
                print(g_id)
                break  # Esco dal ciclo while dopo aver completato con successo la richiesta
            except (GeocoderTimedOut, GeocoderUnavailable):
                time.sleep(1)  # Gestisco il timeout e errore GeocoderUnavailable, e attendo prima di riprovare
```
